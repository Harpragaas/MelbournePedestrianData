```python
import pandas as pd
import pytz
```

```python
#Loading the dataset
monthly_data = pd.read_csv('pedestrian-counting-system-monthly-counts-per-hour.csv')
monthly_data.head()
```

```python
monthly_data['SensingDateTime(Hour)'] = pd.to_datetime(monthly_data['SensingDateTime(Hour)'])
```

```python
# Define the start and end dates for filtering : Last 12 Months
start_date = pd.Timestamp('2023-04-01T00:00:00+10:00')
end_date = pd.Timestamp('2024-05-01T00:00:00+10:00')


pedisterian_data_year = monthly_data[(monthly_data['SensingDateTime(Hour)'] >= start_date) & (monthly_data['SensingDateTime(Hour)'] < end_date)]
```

```python
# Dropping columns
pedisterian_data_year = pedisterian_data_year.drop(columns=['Direction_1','Direction_2'])

#Rename column names
pedisterian_data_year.head()
```

```python
sensor_location = pd.read_csv('pedestrian-counting-system-sensor-locations.csv')
```

```python
# Dropping redundant columns 
sensor_location = sensor_location.drop(columns=['Installation_Date','Note','Direction_1','Direction_2','Latitude','Longitude'])
sensor_location.head()
```

```python
# Merging to get sensor location with Pedestrian counts
ped_count_year_sensor_location = pd.merge(sensor_location,pedisterian_data_year,left_on='Location_ID',right_on='LocationID',how="inner")
ped_count_year_sensor_location.head()
```

```python
ped_count_year_sensor_location = ped_count_year_sensor_location.drop(columns=['Location_x','Sensor_Name_y','LocationID'])

ped_count_year_sensor_location.head()
```

```python
output ='SensorDataLTM.csv'
ped_count_year_sensor_location.to_csv(output,index = False)
```
